1: What is hypothesis testing in statistics?
-->Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves making an assumption (hypothesis) and then using statistical evidence to determine whether to reject or fail to reject it.
It helps evaluate whether observed data significantly differ from what is expected under a given assumption.

2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
-->Null Hypothesis (H₀): Assumes there is no effect or no difference. It represents the default or status-quo assumption.
Example: “The average height of students is 170 cm.”
Alternative Hypothesis (H₁): Suggests that there is an effect or a difference.
Example: “The average height of students is not 170 cm.”
The test determines whether there is enough statistical evidence to reject the null hypothesis in favor of the alternative.

3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
-->The significance level (α) is the probability of rejecting the null hypothesis when it is actually true — i.e., making a Type I error.
Common values are 0.05 or 0.01.
If the p-value ≤ α → Reject H₀ (evidence supports H₁).
If the p-value > α → Fail to reject H₀.

4: What are Type I and Type II errors? Give examples of each.
-->Type I Error (False Positive): Rejecting a true null hypothesis.
Example – A medical test shows a disease when the patient is actually healthy.
Type II Error (False Negative): Failing to reject a false null hypothesis.
Example – A medical test fails to detect a disease that the patient actually has.

5: What is the difference between a Z-test and a T-test? Explain when to use each.
-->Test	When Used	Key Condition
Z-test	When population variance (σ) is known and sample size ≥ 30	Data follows normal distribution
T-test	When population variance is unknown and sample size < 30	Uses sample standard deviation (s)
Both compare sample means, but the T-test accounts for small-sample uncertainty.

```python
#6: Write a Python program to generate a binomial distribution with n = 10 and p = 0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Display mean and variance
print("Mean:", np.mean(data))
print("Variance:", np.var(data))

# Plot histogram
plt.hist(data, bins=range(0, 12), color='lightgreen', edgecolor='black', rwidth=0.9)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#7: mplement hypothesis testing using Z-statistics for a sample dataset in Python.
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu = 50

# Sample statistics
x_bar = np.mean(sample_data)
s = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z = (x_bar - mu) / (s / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Sample Mean = {x_bar:.3f}")
print(f"Z = {z:.3f}, p-value = {p_value:.4f}")
```

```python
#8: Write a Python script to simulate data from a normal distribution and calculate the 95 % confidence interval for its mean.
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Generate data
data = np.random.normal(loc=100, scale=15, size=1000)

# Compute mean and standard error
mean = np.mean(data)
se = st.sem(data)

# 95% Confidence Interval
ci = st.t.interval(confidence=0.95, df=len(data)-1, loc=mean, scale=se)
print(f"Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({ci[0]:.2f}, {ci[1]:.2f})")

# Plot data
plt.hist(data, bins=30, color='lightblue', edgecolor='black')
plt.title("Normal Distribution (μ=100, σ=15)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#9: Write a Python function to calculate Z-scores from a dataset and visualize them using a histogram. Explain what Z-scores represent.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample data
data = np.random.normal(50, 10, 200)

# Compute Z-scores
z_scores = stats.zscore(data)

# Display results
print("First 5 Z-scores:", z_scores[:5])

# Plot histogram
plt.hist(z_scores, bins=25, color='salmon', edgecolor='black')
plt.title("Z-Score Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
